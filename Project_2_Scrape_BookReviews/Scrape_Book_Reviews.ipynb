<h1><center>Scrape Book Reviews</center></h1>

Choose one of your favorite book at goodreads.com (e.g. for the book "The Midninght Library", the URL is: https://www.goodreads.com/book/show/52578297-the-midnight-library?from_choice=true)

- **Write a function to scrape all **reviews on the first page**, including, 
    - **reviewer's name** (see (1) in Figure)
    - **rating** (see (2) in Figure)
    - **date** (see (3) in Figure)
    - **review content** (see (4) in Figure. For each review text, need to get the **complete text**, i.e., need to expand the `more` button. Only text is needed, pictures are not needed. Hint: take a close look at the content of the html file.)
    - **likes** (see (5) in Figure). 
    - If a field, e.g. score, is missing, use `None` to indicate it. 
- `Function Input`: book page URL
- `Function Output`: save all reviews as a DataFrame of columns (`reviewer, rating, date, review, like`). E.g., for the given URL, you can get 30 reviews.(10 points)
- **Note**: GoodReads occasionaly blocks request. You may get an error that is not due to your codes. You may try a couple of times. 
    
    
![alt text](GoodReads.png "GoodReads")

```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
 

def getReviews(page_url):
    url = page_url
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    divs = soup.find_all("div", class_="friendReviews elementListBrown")

    name1 = []
    review1 = []
    rating1 = []
    date1 = []
    like1 = []
    for idx, div in enumerate(divs):
        name = []
        review = []

        p_user = div.select("a.user")
        if p_user == None:
            name = "None"
        else:
            name = p_user[0].get_text()

        p_rate = div.find("span", class_="staticStars notranslate")
        if p_rate == None:
            rate = "None"
        else:
            rate = p_rate.get_text()

        p_date = div.find("a", class_="reviewDate createdAt right")
        if p_date == None:
            date = "None"
        else:
            date = p_date.get_text()

        p_review = div.select("span.readable")
        if p_review == None:
            review = "None"
        else:
            review = p_review[0].get_text()
            review = review[1:]

        p_like = div.find("span", class_="likesCount")
        if p_date == None:
            like = "None"
        else:
            like = p_like.get_text()

        # print(name,review)
        rating1.append((rate))
        name1.append((name))
        date1.append((date))
        review1.append((review))
        like1.append((like))

        c = {"name": name1,
             "rating": rating1,
             "date": date1,
             "review": review1,
             "like": like1}
        reviewr = pd.DataFrame(c)
        reviewr["review"] = reviewr["review"].replace('\n', '', regex=True)
    return reviewr
```

```python
# enter your own url
page_url = 'https://www.goodreads.com/book/show/52578297-the-midnight-library?from_choice=true'

reviews=getReviews(page_url)
reviews
```
